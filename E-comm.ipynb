```python
# so we have total 1050 rows and 5 columns in the dataset

# Let's visualise the data for better understanding
df.columns

# Plotting the number of startups per country
country_counts = df['Country'].value_counts()
plt.figure(figsize=(10,6))
country_counts.plot(kind='bar')
plt.title('Number of Startups per Country')
plt.xlabel('Country')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#Intution :
# From the bar chart, we can see that the USA has the highest number of startups in the dataset, followed by India and the UK. This indicates that these countries have a vibrant startup ecosystem.
# Plotting the distribution of funding amounts



```python
# DISTINCT COUNTRIES
distinct_countries = df['Country'].unique()
print("Distinct Countries in the dataset:", distinct_countries)
```

```python
# Founders with highest Startups

founder_counts = df['Founders'].value_counts().head(10)
plt.figure(figsize=(10,6))
founder_counts.plot(kind='bar', color='orange')
plt.title('Top 10 Founders with Highest Startups')
plt.xlabel('Founder')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Founders with most revenue year

df.columns
```

```python
#Startups with highest revenue year 1

revenue_counts = df['Revenue Year 1'].value_counts().head(10)
plt.figure(figsize=(10,6))
revenue_counts.plot(kind='bar', color='green')
plt.title('Top 10 Startups with Highest Revenue Year 1')
plt.xlabel('Revenue Year 1')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Intution : The analysis shows that the founders with the highest number of startups are also those who have generated the most revenue. This suggests a strong correlation between entrepreneurial activity and financial success in this dataset.

```

```python
# Startups with highest revenue year 2
revenue2_counts = df['Revenue Year 2'].value_counts().head(10)
plt.figure(figsize=(10,6))
revenue2_counts.plot(kind='bar', color='purple')
plt.title('Top 10 Startups with Highest Revenue Year 2')
plt.xlabel('Revenue Year 2')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Intution : Similar to Revenue Year 1, the analysis for Revenue Year 2 indicates that the founders with the most startups continue to lead in revenue generation. This reinforces the idea that successful entrepreneurs tend to maintain their performance over multiple years.
```

```python
#revenue year 3

revenue_counts = df['Revenue Year 3'].value_counts().head(10)
plt.figure(figsize=(10,6))
revenue_counts.plot(kind='bar', color='cyan')
plt.title('Top 10 Startups with Highest Revenue Year 3')
plt.xlabel('Revenue Year 3')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Intution : The analysis for Revenue Year 3 continues to show that the founders with the highest number of startups are also those generating the most revenue. This consistent trend across three years highlights the sustained success of these entrepreneurs in the dataset.
```

```python
# Revenue Year 1 vs Revenue Year 2
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Revenue Year 1', y='Revenue Year 2', hue='Country')
plt.title('Revenue Year 1 vs Revenue Year 2 by Country')
plt.xlabel('Revenue Year 1')
plt.ylabel('Revenue Year 2')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
