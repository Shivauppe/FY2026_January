```python
m = [1,2,3,4,5]
n = [10,20,80,40,150]

plt.plot(m,n,marker = 'o',color = 'g')
plt.title("Line chart to understand the syntax")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(m,n,marker = '*',color = 'b')
plt.title("Line chart to understand the syntax")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(m,n,marker = 'o',color = 'b',linestyle = '--',linewidth = 4)
plt.title("Line chart to understand the syntax")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(m,n,marker = 'o',markersize = 15,color = 'violet',linestyle = '--',linewidth = 4)
plt.title("Line chart to understand the syntax")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [5,1,7,8,2]
y2 = [7,8,3,5,1]

plt.plot(x,y1,marker = 'o',linestyle = '--',linewidth = 4, label = 'Tata motors ')
plt.plot(x,y2,marker = 'o',linestyle = '-',linewidth = 2, label = 'Tata power ')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
x1 = np.linspace(0,10,100)
x2 = np.sin(x1)
```

```python
plt.plot(x1,np.sin(x1),c = 'g',ls = '-',lw = '4',label = 'sinx')
plt.plot(x1,np.cos(x1),c = 'b',ls = '--',lw = '4',label = 'cosx')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x1,np.sin(x1))

plt.subplot(1,2,2)
plt.plot(x1,np.cos(x1),c = 'r')
plt.show()
```
